```python
import pandas
import glob
import os
import re
```

```python
excel_root = "excel todos"
excels = glob.glob( os.path.join(excel_root, '*', '*.xls') )

excel_file = excels[0]
```

```python
def create_index(df, dendrite_type, cell, section, name='synapse_id', provided_sos=None):
    if "Surpass Object" in df:
        sos = df["Surpass Object"]
    elif provided_sos:
        df["Surpass Object"] = provided_sos
        sos = provided_sos
    else:
        raise Exception("No Surpass Object neither provided_sos")
    ids = df["ID"]
    df[name] = (dendrite_type + '-c' + str(cell) + '-s' + str(section) + '-' +
                sos + '-' + ids.apply(lambda x: str(int(x)).zfill(2)))
    df.set_index(name, inplace=True)
    return df
    
```

```python
provided_sos_info= {'colateral':
                        { 6 : {3: 'A'}},
                    'apical':
                        { 
                          5 : {1: 'E'},
                          6 : {1: 'E'}
                        }
                    }
    
def extract_excel_info(excel_file, dendrite_type, cell, section):
    position = pandas.read_excel(excel_file, 'Position', header=1)
    
    provided_sos= provided_sos_info.get(dendrite_type, {}).get(cell, {}).get(section, None)
    
    create_index(position, dendrite_type, cell, section, provided_sos=provided_sos)
    position.rename(columns={'Position X':'x', 
                             'Position Y':'y', 
                             'Position Z':'z',
                             'Surpass Object':'section10um'}, inplace=True)
    
    volume = pandas.read_excel(excel_file, 'Volume', header=1)
    create_index(volume, dendrite_type, cell, section, provided_sos=provided_sos)
    
    area = pandas.read_excel(excel_file, 'Area', header=1)
    create_index(area, dendrite_type, cell, section, provided_sos=provided_sos)
    
    df = position.reindex(columns=['x','y','z','section10um'])
    df['volume'] = volume.Value
    df['area'] = area.Value
     
    df['centroid'] = None
    def f(x):
        x['centroid'] = [x['x'], x['y'], x['z']]
        return x
    df = df.apply(f, axis=1, )
    
    df['cell'] = cell
    df['section'] = section
    df['dendrite_type'] = dendrite_type
    df['source'] = os.path.split(excel_file)[1].split('.')[0]
    df['synapse_id'] = df.index
        
    df = df.drop(['x','y','z'], axis=1)
    return df

#extract_excel_info(excel_file, 'api', 1, 2)
```

```python
def analyze_file_name(excel_file):
    '''
    :return (int cell, str dendrite_type, int section):
    '''
    m = re.match('.*/cel ([0-9]*).*(api|colat).*[0-9]\-([0-9])',excel_file)
    cell, dendrite_type, section = m.groups()
    types = {'api': 'apical','colat':'colateral'}
    dendrite_type = types[dendrite_type]
    return int(cell), dendrite_type, int(section)

analyze_file_name(excel_file)
```

```python
all_synapses = None
for excel_file in excels:
    try:
        cell, dendrite_type, section = analyze_file_name(excel_file)
        df = extract_excel_info(excel_file, dendrite_type, cell, section)

        all_synapses = df if all_synapses is None else all_synapses.append(df)
    except Exception, e:
        print excel_file
        print e
```

```python

sections10 = ['A','B','C','D','E','F']

provided_frame_info = {'apical': { 
                            3 : {1:60},
                            5 : {1:50},
                            6 : {1:50},
                            7 : {1:60}
                           },
                       'colateral': { 
                            0 : {1: 0}
                           }
                       }

def create_distance_frame(df_cell, cell, dendrite_type):
    '''
    This function compute a dict {section: last_distance} 
    With this frame you can know at what distance starts each section 
    depending on previous sections.
    '''
    frame = {0:0}
    
    # Section 1 is special:
    end = provided_frame_info.get(dendrite_type,{}).get(cell,{}).get(1)
    sections = df_cell[df_cell['section'] == 1].section10um.order().unique()
    if end is None:
        end = 50 if len(sections) == 0 else (sections10.index(sections[-1])+1) * 10 
    frame[1] = end
    print 'section {0} has {1} mini sections: {2}'.format(1, len(sections), sections)
    
    # Other sections
    for i in df_cell.section.order().unique():
        if i == 1: continue
        sections = df_cell[df_cell['section'] == i].section10um.order().unique()
        frame[i] = ((sections10.index(sections[-1]) +1) * 10 ) + frame[i-1]
        print 'section {0} has {1} mini sections: {2}'.format(i, len(sections), sections)
    
    return frame

create_distance_frame(all_synapses, 1, 'apical')
```

```python

def compute_dist_section(x, frame):
    return frame[x['section'] - 1] + ((sections10.index(x['section10um']) + 1) * 10)

all_synapses['dist_section'] = None

for dendrite_type in ['apical', 'colateral']:
    df_typed = all_synapses[all_synapses.dendrite_type == dendrite_type]
    for cell in df_typed.cell.order().unique():
        print dendrite_type, cell
        df_cell = df_typed[df_typed['cell']==cell]
        frame =  create_distance_frame(df_cell, cell, dendrite_type)
        print frame
        df_cell['dist_section'] = df_cell.apply(compute_dist_section, axis=1, args=(frame,))
        all_synapses.update(df_cell)
        
        
```

```python
vrml_root = "vrml todos PSD"
vrmls = glob.glob( os.path.join(vrml_root, '*', '*.vrml') )

```

```python
def get_df_files(files):
    files = pandas.Series(files)
    
    s = files.apply(analyze_file_name)
    df = pandas.DataFrame(s.tolist(), columns=['cell', 'dendrite_type', 'section'])
    
    #cell, dendrite_type, section = spo
    
    filename = files.str.split('/').apply(lambda x: x[2]).str.split('.').apply(lambda x: x[0])
    
    df['filename'] = filename
    df.set_index(['cell', 'dendrite_type', 'section'], inplace=True)
    return df
```

```python
df_excels = get_df_files(excels)
df_excels.rename(columns={'filename':'excel'}, inplace=True)
df_vrmls = get_df_files(vrmls)
df_vrmls.rename(columns={'filename':'vrml'}, inplace=True)
```

```python
ferets = glob.glob( os.path.join(vrml_root, '*', '*.csv') )
```

```python
def get_df_ferets(files):
    df_ferets = None
    for filename in files:
        cell, dendrite_type, section = analyze_file_name(filename)
        df = pandas.read_csv(filename)
        df['cell'] = cell
        df['dendrite_type'] = dendrite_type
        df['section'] = section
        df['Object'] = df.Object.apply(lambda x: str(int(x)).zfill(2))
    
        df['centroid'] = None
        def f(x):
            x['centroid'] = [x['X'], x['Y'], x['Z']]
            return x
        df = df.apply(f, axis=1)
        
        df = df.rename(columns={'X':'x','Y':'y','Z':'z'})
        
        df_ferets = df_ferets.append(df) if df_ferets else df
    
    return df_ferets

df_ferets = get_df_ferets(ferets)

assert len(all_synapses) == len(df_ferets), 'No same number of synapses and ferets'
```
